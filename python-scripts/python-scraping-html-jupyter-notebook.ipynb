```python
import requests as re 
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
# I just learned that the "re" is an alias for the requests library. Learned that from this video: https://www.youtube.com/watch?v=YnIfMZingMA
```

```python
r = re.get('https://docs.oracle.com/en/database/oracle/oracle-database/21/sqlqr/SQL-Statements.html')
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
# print(soup.body)
```

```python
# print(soup.find_all('div', class_='section'))

body = soup.find_all('div', class_='section')
```

```python
print(body)
```

```python
sqlstatement = []
sqlstatement.clear()
for i in soup.find_all('p', class_='subhead2'):
    sqlstatement.append((i.find('a', {'target': '_blank'})).text)
print(sqlstatement)
```

```python
syntaxexample = []
syntaxexample.clear()
for j in soup.find_all('pre', class_='oac_no_warn'):
    syntaxexample.append(j.text)
# print(syntaxexample)
```

```python
df = pd.DataFrame(
    {'sqlstatement':sqlstatement,
    'syntaxexample':syntaxexample
    }
)
```
